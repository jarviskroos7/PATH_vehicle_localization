#### Normalize GPS coordinate to 2D plannar axes

```python
def next_coord_conversion(coord1, coord2, coord_prev=[0, 0]):

    x0 = coord_prev[0]
    y0 = coord_prev[1]
    lat1 = coord1[0] # x
    lon1 = coord1[1] # y
    lat2 = coord2[0]
    lon2 = coord2[1]

    geodesic = pyproj.Geod(ellps='WGS84')
    bearing, inv_bearing, distance = geodesic.inv(lon1, lat1, lon2, lat2)

    coord_next = [
        x0 + distance * math.cos(bearing * math.pi / 180), 
        y0 + distance * math.sin(bearing * math.pi / 180)
    ]

    return coord_next
```

```python
# runtime ~= 30s

ll_coord_next_lst = []
rl_coord_next_lst = []

# left lane
for index, row in ll.iterrows():
    
    coord1 = [row['y'], row['x']]
    
    if index == 0:
        coord2 = [ll.iloc[index+1]['y'], ll.iloc[index+1]['x']]
        coord_next = next_coord_conversion(coord1, coord2)
    elif index == ll.shape[0]-1:
        pass
    else:
        coord2 = [ll.iloc[index+1]['y'], ll.iloc[index+1]['x']]
        coord_prev = ll_coord_next_lst[-1]
        coord_next = next_coord_conversion(coord1, coord2, coord_prev)
    
    ll_coord_next_lst.append(coord_next)

# right lane
for index, row in rl.iterrows():
    
    coord1 = [row['y'], row['x']]
    
    if index == 0:
        coord2 = [rl.iloc[index+1]['y'], rl.iloc[index+1]['x']]
        coord_next = next_coord_conversion(coord1, coord2)
    elif index == rl.shape[0]-1:
        pass
    else:
        coord2 = [rl.iloc[index+1]['y'], rl.iloc[index+1]['x']]
        coord_prev = rl_coord_next_lst[-1]
        coord_next = next_coord_conversion(coord1, coord2, coord_prev)
    
    rl_coord_next_lst.append(coord_next)
```

## Ground Truth Segment Generation

### [v2] Fixed length, random start
- generate 500 ground truth segments from the surveyed test track (center lane)
- each ground truth segments is around 200m in length, with 100*standard normal distribution
- coordinates are normalized to 0-1
- no noise and sub-sampling are performed at this step
- training and validation set: left lane
- testing: right lane

```python
def seg_generation(traj_truth_df, seg_size, seg_size_sig, num_samples=500, seed=0):
    
    # ensure reproducibility
    np.random.seed(seed)
    seg_gt = []

    for x in np.random.randint(traj_truth_df.shape[0]-seg_size, size=num_samples):
        seg_size += round(seg_size_sig * randn())
        # seg_sample.append(traj_truth_df[x:x+seg_size, :])
        seg_gt.append(traj_truth_df.iloc[x:x+seg_size])

    return seg_gt
```

```python
tot_dist = ll_raw.iloc[-1]['Distance horizontal (m)']
seg_dist = 200 # [m]
seg_size = round(seg_dist / (tot_dist / ll.shape[0])) # convert to the length of sample
seg_size_sig = 0 # 100 # length variance

# save all columns
ll_seg_gt = seg_generation(ll, seg_size, seg_size_sig)
rl_seg_gt = seg_generation(rl, seg_size, seg_size_sig)
# ll_seg_gt = seg_generation(ll, seg_size, seg_size_sig)
# rl_seg_gt = seg_generation(rl, seg_size, seg_size_sig)

# save only GPS coordinate columns
gps_colns = ['y', 'x']# , 'y_norm', 'x_norm']
ll_seg_gps_gt = seg_generation(ll[gps_colns], seg_size, seg_size_sig)
rl_seg_gps_gt = seg_generation(ll[gps_colns], seg_size, seg_size_sig)
```

### [v4] Fixed start, random length
- generate 500 ground truth segments from the surveyed test track
- each segment would start from the starting point and have a random length, with maximum length being the entire track 
- no noise and sub-sampling are performed at this step
- training and validation set: left lane
- testing: right lane

```python
def seg_generation_v4(track, min_seg_size, num_samples=250, seed=0):
    
    # ensure reproducibility
    np.random.seed(seed)
    colns = ['x', 'y', 'theta', 'Vx', 'omega_Zv', 'alpha_Xv', 'Distance horizontal (m)']
    track = track[colns]
    seg_gt = []
    start_idx = 0

    # append to list of arrays
    for end_idx in np.random.randint(min_seg_size, track.shape[0], size=num_samples):
        seg_gt.append(track.iloc[start_idx:end_idx])

    return seg_gt
```

## Noisy Sampled Segment Generation

#### add_gps_noise_downsample()

- A variant of this function can be found in `src/util.py`

```python
def add_gps_noise_downsample(seg_samples, sampling_rate, noise_perc, seed=0):
    """
    seg_samples: list of df, ll_seg_samples
    sampling_rate: unit [hz]
    noise_perc: 0-1, fraction
    """

    # known: the sample rate of the RTK-GPS unit is actually 100hz, not 250hz
    rtk_sampling_rate = 100 # 250
    sampling_spacing = round(rtk_sampling_rate / sampling_rate)
    seg_samples_noise = []

    try: 
        np.random.seed(seed)
        for seg_sample in seg_samples:
            seg_sample = seg_sample.iloc[::sampling_spacing]
            seg_sample['y_noise'] = seg_sample['y'] + noise_perc * randn(*seg_sample['y'].shape)
            seg_sample['x_noise'] = seg_sample['x'] + noise_perc * randn(*seg_sample['x'].shape)
            seg_samples_noise.append(seg_sample)
    
    # [v5] full track, 50 samples
    except:
        
        for i in np.arange(50):
            rng = np.random.RandomState(i)
            seg_sample = seg_samples.iloc[::sampling_spacing]
            seg_sample['y_noise'] = seg_sample['y'] + noise_perc * rng.randn(*seg_sample['y'].shape)
            seg_sample['x_noise'] = seg_sample['x'] + noise_perc * rng.randn(*seg_sample['x'].shape)
            seg_samples_noise.append(seg_sample)
        
    return seg_samples_noise
```

```python
def add_gps_noise_downsample_gpd(seg_samples, sampling_rate, noise_perc, seed=0):
    """
    seg_samples: list of df, ll_seg_samples
    sampling_rate: unit [hz]
    noise_perc: 0-1, fraction
    """

    np.random.seed(seed)
    # known: the sample rate of the RTK-GPS unit is 250hz
    rtk_sampling_rate = 250
    sampling_spacing = round(rtk_sampling_rate / sampling_rate)
    seg_samples_gps = []

    for seg_sample in seg_samples:
        seg_sample = seg_sample.iloc[::sampling_spacing]
        seg_sample['y_noise'] = seg_sample['y'] + noise_perc * randn(*seg_sample['y'].shape)
        seg_sample['x_noise'] = seg_sample['x'] + noise_perc * randn(*seg_sample['x'].shape)
        seg_samples_gps.append(seg_sample)

    return seg_samples_gps
```

```python
# downsample from 250hz to 1hz, adding 1% spatial noise

ll_seg_gps_noise = add_gps_noise_downsample_gpd(ll_seg_gps_gt, 10, 0.0001)
rl_seg_gps_noise = add_gps_noise_downsample_gpd(rl_seg_gps_gt, 10, 0.0001)

ll_seg_noise = add_gps_noise_downsample_gpd(ll_seg_gt, 10, 0.0001)
rl_seg_noise = add_gps_noise_downsample_gpd(rl_seg_gt, 10, 0.0001)
```

## Noisy Lane Marking Segments Generation

```python
ll_marks = pd.read_csv('../data/acm_track/augmented_lane_marking/ll_lane.csv.gzip', compression='gzip', index_col=0)
rl_marks = pd.read_csv('../data/acm_track/augmented_lane_marking/rl_lane.csv.gzip', compression='gzip', index_col=0)

ll_marks = ll_marks[['y', 'x', 'Vx', 'left_mark_lat', 'left_mark_lon', 'right_mark_lat', 'right_mark_lon', 'Distance horizontal (m)']]
rl_marks = rl_marks[['y', 'x', 'Vx', 'left_mark_lat', 'left_mark_lon', 'right_mark_lat', 'right_mark_lon', 'Distance horizontal (m)']]
```

```python
# minimum segment length to be around 100m
num_samples = 250
seg_length_avg = 1000 # [m]
seg_length_std = 200 # [m]

tot_dist = max(ll_marks.iloc[-1]['Distance horizontal (m)'], rl_marks.iloc[-1]['Distance horizontal (m)'])
seg_size_avg = round(seg_length_avg / (tot_dist / max(ll_marks.shape[0], ll_marks.shape[0]))) # convert to the length of sample
seg_size_std = round(seg_length_std / (tot_dist / max(ll_marks.shape[0], ll_marks.shape[0]))) # convert to the length of sample

ll_marks_seg = utils.seg_generation(
    ll_marks, seg_size_avg, seg_size_std, num_samples=num_samples, ver=1
)
len(ll_marks_seg), ll_marks_seg[0].shape
```

```python
# left lane

ll_seg_sampled_noise = []
noise_perc = 0.000005
colns = ['left_mark_lat', 'left_mark_lon', 'right_mark_lat', 'right_mark_lon']

for i in range(num_samples):
    seg = ll_marks_seg[i]

    # downsampling from 100 to 10hz
    ll_sampled_noise = utils.sample_gps(seg, sampling_rate=10)

    # augment spatial noise, 5-decimal places, ~ 1m error    
    ll_sampled_noise = utils.gps_noise_augmentation(ll_sampled_noise, noise_perc, colns, seed=i)

    # save as numpy array for space saving
    ll_seg_sampled_noise.append(ll_sampled_noise) #.to_numpy())

print(len(ll_seg_sampled_noise), ll_seg_sampled_noise[0].shape)
# ll_seg_sampled_noise = np.array(ll_seg_sampled_noise)

with gzip.open('../data/acm_track/augmented_lane_marking/ll_lane_seg_250_sampled_noise_v1.pkl.gzip', 'wb') as f:
    pickle.dump(ll_seg_sampled_noise, f)
```
